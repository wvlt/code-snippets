```python
from matplotlib import pyplot as plt
from matplotlib import image as mpimg
```

```python
stinkbug_image = mpimg.imread('stinkbug.png')
```

```python
stinkbug_image.shape
```

1. RGB: Red, Green and Blue
2. RGBA: Red, Green, Blue and Alpha

```python
plt.figure(figsize=(8,8))
plt.imshow(stinkbug_image)
```

```python
stinkbug_image_red = stinkbug_image[:, :, 0]
```

```python
plt.figure(figsize=(8,8))

plt.imshow(stinkbug_image_red, cmap='Greens')
plt.title('Red Channel')
plt.colorbar();
```

```python
plt.hist(stinkbug_image_red.ravel(), bins=256);
```

```python
plt.imshow(stinkbug_image_red, clim=(0, 0.4))
```

```python
fig = plt.figure(figsize=(14,10))
ax = fig.add_subplot(1, 2, 1)
imgplot = plt.imshow(stinkbug_image)
ax.set_title('Before')
plt.colorbar(ticks=[0.1, 0.3, 0.5, 0.7], orientation='horizontal')

ax = fig.add_subplot(1, 2, 2)
imgplot = plt.imshow(stinkbug_image)
imgplot.set_clim(0, 0.4)
ax.set_title('After')
plt.colorbar(ticks=[0.0, 0.1, 0.2, 0.3, 0.4], orientation='horizontal')
```
