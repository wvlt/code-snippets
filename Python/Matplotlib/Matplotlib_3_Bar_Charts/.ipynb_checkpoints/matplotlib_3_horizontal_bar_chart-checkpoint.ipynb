```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data.csv', index_col=[0])
```

```python
data.head()
```

```python
plt.figure(figsize=(8,8))

data = data.sort_values('Frequency')

plt.barh(data['Language'], data['Frequency'])

plt.title('Number of Responses for Each Programming Language')
plt.ylabel('Programming Language')
plt.xlabel('Frequency')
plt.grid(alpha=0.4)

plt.show()
```

```python
fig, axes = plt.subplots(1,2, figsize=(14,7))

data = data.sort_values('Frequency')
axes[0].barh(data['Language'], data['Frequency'])

data = data.sort_values('Frequency', ascending=False)
axes[1].barh(data['Language'], data['Frequency'], color='red')

plt.subplots_adjust(wspace=0.4)

plt.show()
```
