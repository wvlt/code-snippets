```python
mean_python = np.mean(python_salary)
mean_java = np.mean(java_salary)
```

```python
fig, axes = plt.subplots(1,2,figsize=(12,7))

bins = [10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000, 130000]

axes[0].hist(python_salary, bins=bins, edgecolor='black', color='red')
axes[0].axvline(mean_python, color='green', label='Mean Python Salary', linewidth=4)


axes[1].hist(java_salary, bins=bins, edgecolor='black', color='blue')
axes[1].axvline(mean_java, color='green', label='Mean Java Salary', linewidth=4)

axes[0].legend()
axes[0].set_title('Python Developer Pay Distribution in USD', fontsize=16)
axes[0].set_xlabel('Median Pay in USD')
axes[0].set_ylabel('Number of Respondants')
axes[0].grid(alpha=0.25)

axes[1].legend()
axes[1].set_title('Java Developer Pay Distribution in USD', fontsize=16)
axes[1].set_xlabel('Median Pay in USD')
axes[1].set_ylabel('Number of Respondants')
axes[1].grid(alpha=0.25)

plt.show()
```
