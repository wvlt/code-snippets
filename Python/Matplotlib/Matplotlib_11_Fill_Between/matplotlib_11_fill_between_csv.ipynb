```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
x = data['Age']
py_salary = data['Python']
java_salary = data['JavaScript']
```

```python
plt.figure(figsize=(7,7))

plt.plot(x, py_salary, label='Python Developer Salaries')
plt.plot(x, java_salary, label='Java Developer Salaries')

plt.fill_between(x, py_salary, java_salary, where=py_salary>java_salary, interpolate=True,
                 color='grey', alpha=0.4, label='Python Earning More Than Java')

plt.fill_between(x, py_salary, java_salary, where=py_salary<java_salary, interpolate=True,
                 color='red', alpha=0.4, label='Java Earning More Than Python')

plt.title('Fill Between Python Salaries and Java Salaries')
plt.xlabel('Age')
plt.ylabel('Salary(USD)')
plt.grid(alpha=0.4)
plt.legend()
plt.show()
```
