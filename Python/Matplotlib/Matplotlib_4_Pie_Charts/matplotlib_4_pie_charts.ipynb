```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
pets = [3, 2, 5, 1]
labels = ['Cats', 'Dogs', 'Birds', 'Rabbit']
colors = ['red', 'blue', 'yellow', 'green']
```

```python
plt.figure(figsize=(7,7))

plt.pie(pets, colors=colors, labels=labels, wedgeprops={'edgecolor': 'black'})

plt.title('My Awesome Pets')

plt.show()
```

```python
data = pd.read_csv('data.csv', index_col=[0])
```

```python
data = data.sort_values('Frequency', ascending=False)
```

```python
values = data['Frequency']
```

```python
labels = data['Language']
```

```python
labels[:7]
```

```python
explode = [0, 0, 0, 0.2, 0, 0, 0]
```

```python
plt.figure(figsize=(7,7))

plt.pie(values[:7], labels=labels[:7], explode=explode, shadow=True, startangle=180,
        autopct='%1.1f%%',
        wedgeprops={'edgecolor':'white'})

plt.title('Popularity of Programming Langauges', fontsize=16)

plt.show()
```
