```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = np.array([1, 3, 4, 5, 7])
y = np.array([1, 9, 16, 25, 49])
```

```python
plt.figure(figsize=(7,7))

plt.plot(x,y, 'red')
plt.step(x,y, '^-g', where='pre')
plt.bar(x,y)

plt.title('Awesome Step Plot')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.grid(alpha=0.4)
plt.show()
```
