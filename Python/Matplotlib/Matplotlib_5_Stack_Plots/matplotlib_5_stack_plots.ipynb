```python
from matplotlib import pyplot as plt
```

```python
# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

amir = [8, 6, 6, 6, 7, 5, 5, 5, 4, 4, 3, 3]
james = [0, 0, 1, 2, 1, 0, 1, 3, 2, 1, 2, 0]
callum = [0, 2, 1, 0, 0, 3, 2, 0, 2, 3, 3, 5]

labels=['Amir', 'James', 'Callum']
colors=['orange', 'green', 'blue']
```

```python
plt.figure(figsize=(12,7))

plt.stackplot(months, amir, james, callum, labels=labels, colors=colors)

plt.title('Share of Project', fontsize=16)
plt.xlabel('Month of the year')
plt.ylabel('Total Share')
plt.legend(loc=9)
plt.grid(alpha=0.6)
plt.show()
```
