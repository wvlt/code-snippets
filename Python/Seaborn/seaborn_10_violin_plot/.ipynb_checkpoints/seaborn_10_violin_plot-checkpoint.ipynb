```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.catplot(data=tips, y='tip', x='sex', hue='smoker', split=True,
            kind='violin', inner='box')
```

```python
sns.violinplot(data=tips, y='total_bill', x='day', hue='smoker', split=True,
            inner='stick', palette='Greens')
```
