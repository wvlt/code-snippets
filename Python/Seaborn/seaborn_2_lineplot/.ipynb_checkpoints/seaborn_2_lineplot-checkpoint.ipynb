```python
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
sns.lineplot(data=flights, x='year', y='passengers')
```

```python
feb_flights = flights.query("month == 'Feb'")
```

```python
sns.lineplot(data=feb_flights, x='year', y='passengers')
```

```python
flights.head()
```

```python
plt.figure(figsize=(7,7))
sns.lineplot(data=flights, x='year', y='passengers', hue='month', style='month')
```

```python
plt.figure(figsize=(7,7))

sns.relplot(data=flights, x='year', y='passengers', row='month',
            hue='month', kind='line', style='month')
plt.title('Number of Passengers Over Years')
plt.grid(alpha=0.3);
```
