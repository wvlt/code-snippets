```python
import seaborn as sns
import numpy as np
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
rug_kws = {'color': 'red', 'linewidth':0.25}
sns.displot(data=tips, x='total_bill', kind='ecdf', rug=True, rug_kws=rug_kws, linewidth=4)
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex')
sns.rugplot(data=tips, x='total_bill', hue='sex', height=0.05, legend=False)
sns.rugplot(data=tips, y='tip', height=0.05)
```
