```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
sns.displot(data=penguins, x='body_mass_g', kind='hist', kde=True)
```

```python
sns.displot(data=penguins, x='flipper_length_mm', col='species', kde=True,
            hue='species', bins=30)
```

```python
plt.figure(figsize=(7,7))
sns.histplot(data=penguins, x='body_mass_g', y='flipper_length_mm', hue='species', palette='turbo')
```
